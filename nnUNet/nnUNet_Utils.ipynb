```python
import os
import numpy as np

def verificar_shape_npz(directory):
    archivos_con_shape_incorrecto = []
    count = 0
    for filename in os.listdir(directory):
        if filename.endswith(".npz"):
            print(count)
            filepath = os.path.join(directory, filename)
            try:
                data = np.load(filepath)
                shape_data = data['data'].shape
                shape_seg = data['seg'].shape
                if (shape_data[0] != 3 | shape_seg[0] != 3) :
                    archivos_con_shape_incorrecto.append(filename)
                    print("Archivo erroneo encontrado!")
            except Exception as e:
                print(f"Error al procesar el archivo {filename}: {e}")
            count += 1
    return archivos_con_shape_incorrecto

directorio = "/content/drive/MyDrive/TFG/V3/nnUNet_preprocessed/Dataset212_CholecSeg8kV2/nnUNetPlans_2d"  # Reemplaza con la ruta a tu directorio
archivos_incorrectos = verificar_shape_npz(directorio)

if archivos_incorrectos:
    print("Los siguientes archivos tienen un primer campo de shape distinto de 3 en 'data':")
    for archivo in archivos_incorrectos:
        print(archivo)
else:
    print("Todos los archivos tienen un primer campo de shape igual a 3 en 'data'.")
```
